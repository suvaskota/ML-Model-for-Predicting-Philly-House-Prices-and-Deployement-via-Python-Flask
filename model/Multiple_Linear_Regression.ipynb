```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("philly_housing.csv")
df
```

```python

df.columns.values
```

```python
df_initial = df[['Zillow Estimate', 'yearBuilt', 'finished \n(SqFt)', ' bathrooms ', ' bedrooms ', 'PropType']]
```

```python
df_cleaned = df_initial[pd.notnull(df_initial[['Zillow Estimate', 'yearBuilt', 'finished \n(SqFt)', ' bathrooms ', ' bedrooms ', 'PropType']])]
```

```python
df_cleaned.dropna(axis=0)
```

```python
df_cleaned = df_cleaned.dropna(axis=0)
```

```python
df_cleaned.to_csv("cleaned_data.csv")
```

```python
df_cleaned = df_cleaned.replace(' -   ', np.nan)
```

```python
df_cleaned = df_cleaned.dropna(axis=0)
```

```python
df_cleaned['Zillow Estimate'] = df_cleaned['Zillow Estimate'].str.replace(',', '')
```

```python
df_cleaned.to_csv("cleaned_data.csv")
```

```python
y = df_cleaned['Zillow Estimate']
x = df_cleaned[['yearBuilt', 'finished \n(SqFt)', ' bathrooms ', ' bedrooms ']]
```

```python
linear_regression = LinearRegression(normalize=True)

```

```python
linear_regression.fit(x, y)
```

```python
data = {'yearBuilt' : [2001], 'finished \n(SqFt)': [2000], ' bathrooms ': [2.5], ' bedrooms ': [4]}
```

```python
test = pd.DataFrame.from_dict(data)
```

```python
test
```

```python
print(linear_regression.predict(test))
```

```python
def predict_price(year, sqFt, bathrooms, bedrooms):
    data = {'yearBuilt' : [year], 'finished \n(SqFt)': [sqFt], ' bathrooms ': [bathrooms], ' bedrooms ': [bedrooms]}
    prediction = pd.DataFrame.from_dict(data)
    return linear_regression.predict(prediction)[0]
    
```

```python
predict_price(2005, 4000, 3, 5)
```
